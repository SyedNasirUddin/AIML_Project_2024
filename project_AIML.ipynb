1\. Handle Missing Values: Check for missing values in each column and

missed values with a mean and median a numerical columns and the mode
for categorically columns.

2\. Convert Datatypes: Ensure that each column has the correct data
type. For example, age, BMI, and expenses should be numeric, while sex,
smoking, and region should be categorical.

3\. Handle Categorical Data: Encode category variables using one-hot
encoding or label encoding. This step is necessary for algorithms that
cannot handle categorical data directly.

4\. Normalize/Standardize Numerically Data Methods: Normalize or types
of standard no. of features to ensure that they're on a similar scales.
This step is crucial for models easier to scale of features, same as KNN
or neural networks.

5\. Remove Outliers: Identify and remove outliers from numerical
features if necessary. Outliers can repeatedly impact in the
performances of some AIML learning algorithms.

6\. Feature Scaling: Scale the features if needed. For example, you
might use Min-Max scaling or standard scaling to bring all features to
the same scale.

7\. Feature Engineering (Model): Add new types from present ones if you
believe it will improve model performances. For real example, you might
make a feature representing them as interaction between ages.

8\. Testing Sets of Data: Divide the datasets where trainings and
testing’s data sets to correct these performances of the model. Typical
splits including 70-30 or 80-20 to use in train purpose and testing
purpose, simultaneously.

9\. Handle Class Imbalance: If your dataset has class imbalance (e.g.,
significantly more instances of one class than another), consider
techniques such as oversampling, under sampling, or using algorithms
that handle class imbalance well.

10\. Verifying Data Integrity: Double-check that the pre-processed data
maintains integrity and consistency. Ensure that the pre-processing
steps have not introduced any errors or unexpected behaviour.
